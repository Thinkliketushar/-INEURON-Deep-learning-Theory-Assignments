#### 3.	If you want to build a deep sequence-to-sequence RNN, which RNN layers should have return_sequences=True? What about a sequence-to-vector RNN?

<code>When building a deep sequence-to-sequence RNN (Recurrent Neural Network), you typically have multiple layers, and you would set return_sequences=True for all the recurrent layers. This is necessary to ensure that each RNN layer in the stack produces an output sequence that can be used as input for the next layer in the stack.
In contrast, for a sequence-to-vector RNN, you typically set return_sequences=False (or you can omit it because it's often the default behavior). This configuration means that the final output of the RNN will be a single vector summarizing the entire input sequence.
